#Visualización con Matplotlib y Seaborn

#Librería Matplotlib

Se importa con la siguiente instrucción

```python
import matplotlib.pyplot as plt
```

Cuando creamos un gráfico y deseamos mostrarlo en pantalla debemos usar plt.show()

```python
import numpy as np
x = np.linspace(0, 10, 100)  #Creamos array de 100 puntos entre 0 y 10

fig = plt.figure()     #Definimos una figura
plt.plot(x, np.sin(x), '-')  #Representamos la función sen(x) con línea discontinua
plt.plot(x, np.cos(x), '--')  #Representamos la función cos(x) con línea continua

plt.show()
```


Para todas las gráficas de Matplotlib comenzamos creando una figura y los ejes. En su forma más simple, se pueden crear una figura y ejes de la siguiente manera:

```python
fig = plt.figure()
ax = plt.axes()
```

Y añadimos los datos a representar

```python
plt.plot(x,np.sin(x))
```

Podemos insertar título y nombres a los ejes

```python
plt.plot(x,np.sin(x))
plt.title("Función seno")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

Podemos elegir el color y el tipo de línea, así como insertar una leyenda si hemos definido las etiquetas de los distintos datos representados

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')

plt.title("Funciones seno y coseno")
plt.xlabel("x")

plt.legend();
```

Para representar sólo puntos podemos usar marker '.' o 'o'

```python
plt.plot(x, np.sin(x), 'o')
```

O podemos usar plt.scatter, que dispone de más posibilidades

```python
plt.scatter(x, np.sin(x), marker='o');
```

```python
plt.scatter(x, np.sin(x), color='green', marker='.', s=10*x)  #El tamaño del punto varía 
```

Para representar histogramas (gráficos de frecuencias) usamos hist()

```python
data = np.random.rand(1000) #Creamos un array de 1000 puntos aleatorios entre 0 y 1

plt.hist(data) #Representamos el histograma (frecuencia o número de veces que aparece cada número)
```

```python
data = np.random.randn(1000) #Creamos un array de 1000 puntos con distribución normal estándar

plt.hist(data)
```

Se puede indicar el número de intervalos o bins

```python
plt.hist(data, bins=30)
plt.show()
```

Y es posible añadir varios parámetros más para enriquecer la visualización

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

##Gráficos en 3D

```python
fig = plt.figure()
ax=plt.axes(projection='3d')
```

```python
ax = plt.axes(projection='3d')

# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(40, 35)
```

#Librería Seaborn

Permite algunas visualizaciones muy útiles para el análisis exploratorio de datos y compatible con la librería Pandas

```python
import seaborn as sns
sns.set()
plt.style.use('seaborn')
```

```python
#Creamos los datos a representar
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)
```

```python
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');
```

Representar histogramas

```python
import pandas as pd
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])

for col in 'xy':
    plt.hist(data[col], alpha=0.5)
```

Representar histograma con densidad de probabilidad

```python
sns.distplot(data['x'])
```

Los datos de un DataFrame pueden visualizarse fácilmente con Seaborn y mostrar características estadísticas como correlaciones entre variables

Importamos un dataset denominado Iris con las medidas realizadas con flores de distintas especies. Las medidas corresponden a la longitud y anchura de los pétalos y los sépalos de las flores 

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
c=sns.FacetGrid(iris,col='species')
c.map(plt.scatter,'sepal_length','sepal_width')
```

```python
sns.boxplot(x='species',y='sepal_length',data=iris)
```

```python
sns.violinplot(x='species',y='sepal_length',data=iris)
```

```python
sns.swarmplot(x='species',y='sepal_length',data=iris)
```

##Pair Plots
Los Pair plots son muy útiles para explorar correlaciones entre datos multidimensionales, cuando queremos trazar todas las combinaciones de variables entre sí.

```python
sns.pairplot(iris, hue='species', size=2.5);
```
